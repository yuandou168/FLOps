```python
import matplotlib.pyplot as plt

# Data for the plot
experiments = ['exp1', 'exp2', 'exp3', 'exp4', 'exp5', 'exp6', 'exp7', 'exp8', 'exp9', 'exp10']
time_in_minutes = [10.29, 9.80, 9.77, 9.75, 9.71, 9.65, 10.00, 9.42, 9.96, 9.51]
test_accuracy = [98.14, 98.22, 98.27, 98.13, 98.16, 98.01, 98.13, 98.14, 97.96, 98.11]

font = {'family': 'serif', 'weight': 'normal', 'size': 12}
plt.rc('font', **font)

# Re-plotting with test accuracy line for the second y-axis
fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:cyan'
# Plotting Time in Minutes as bars
ax1.bar(experiments, time_in_minutes, color=color)
ax1.set_ylabel('Time in Minutes (mins)', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(0, 16)

# Adding labels to the bars
for i, value in enumerate(time_in_minutes):
    ax1.text(i, value + 0.1, f'{value:.2f}', ha='center', color=color, fontsize=10)

# Adding gridlines for the left y-axis (time)
ax1.grid(True, linestyle='--', which='both', axis='y', color='gray', alpha=0.6)

# Adding mean line for Time in Minutes
mean_time = sum(time_in_minutes) / len(time_in_minutes)
ax1.axhline(y=mean_time, color='tab:purple', linestyle='--', linewidth=1)
# ax1.text(len(experiments)-0.5, mean_time + 0.2, f'Mean Time: {mean_time:.2f}', color='tab:purple')

# Create second axis for Test Accuracy
ax2 = ax1.twinx()
color = 'tab:orange'
ax2.plot(experiments, test_accuracy, color=color, marker='*', markersize=10, linestyle='-', linewidth=1.5)
ax2.set_ylabel('Test Accuracy (%)', color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(90, 100)

# Adding labels to the test accuracy points
for i, value in enumerate(test_accuracy):
    ax2.text(i, value - 0.4, f'{value:.2f}%', ha='center', color=color)

# Adding gridlines for the right y-axis (accuracy)
ax2.grid(True, linestyle='--', which='both', axis='y', color='gray', alpha=0.6)

# Adding the line for Test Accuracy mean
mean_accuracy = sum(test_accuracy) / len(test_accuracy)
ax2.axhline(y=mean_accuracy, color="tab:pink", linestyle='--', linewidth=1)
# ax2.text(len(experiments)-0.5, mean_accuracy - 0.5, f'Mean Accuracy: {mean_accuracy:.2f}%', color=color)

# Customizing the layout and displaying the plot with a test accuracy line and time mean line
plt.title('FL Training Time and Test Accuracy for Various Cloud Configurations')
plt.tight_layout()
plt.show()
```

```python
# Re-plotting with the mean value of the accuracy label moved above the arrow
fig, ax1 = plt.subplots(figsize=(8, 6))

round_numbers = list(range(1, 11))  # Explicit round numbers for x-axis

font = {'family': 'serif',
        'weight': 'normal',
        'size': 12}
plt.rc('font', **font)
color = 'teal'
# Plotting Time in Minutes as bars
ax1.bar(round_numbers, time_in_minutes, color=color, width=0.4)
ax1.set_ylabel('Time in Minutes (mins)', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(0, 18)
ax1.set_xlabel('Experiment Rounds')
ax1.set_xticks(round_numbers)  # Set the round numbers explicitly on x-axis

# Adjusting the labels to be a little higher above the bars
for i, value in enumerate(time_in_minutes):
    ax1.text(i + 1, value + 0.5, f'{value:.2f}', ha='center', color=color, fontsize=12)

# Adding gridlines for the left y-axis (time)
ax1.grid(True, linestyle='--', which='both', axis='y', color='gray', alpha=0.6)

# Adding mean line for Time in Minutes
mean_time = sum(time_in_minutes) / len(time_in_minutes)
ax1.axhline(y=mean_time, color=color, linestyle='--', linewidth=1)

# Adding arrow to point to the mean of time in minutes
ax1.annotate(f'Mean Time: {mean_time:.2f} mins', xy=(5.5, mean_time), xytext=(5.5, mean_time + 2),
             arrowprops=dict(facecolor=color, shrink=0.05, width=0.5), fontsize=12, color=color)

# Create second axis for Test Accuracy
ax2 = ax1.twinx()
color = 'indianred'
ax2.plot(round_numbers, test_accuracy, color=color, marker='*', markersize=10, linestyle='-', linewidth=1.5)
ax2.set_ylabel('Test Accuracy (%)', color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(90, 100)

# Adding labels to the test accuracy points (above the points)
for i, value in enumerate(test_accuracy):
    ax2.text(i + 1, value + 0.3, f'{value:.2f}', ha='center', color=color)

# Adding gridlines for the right y-axis (accuracy)
ax2.grid(True, linestyle='--', which='both', axis='y', color='gray', alpha=0.6)

# Adding the line for Test Accuracy mean
mean_accuracy = sum(test_accuracy) / len(test_accuracy)
ax2.axhline(y=mean_accuracy, color=color, linestyle='--', linewidth=1)

# Moving the mean accuracy annotation above the arrow
ax2.annotate(f'Mean Accuracy: {mean_accuracy:.2f}%', xy=(5.5, mean_accuracy), xytext=(5.5, mean_accuracy + 1.2),
             arrowprops=dict(facecolor='indianred', shrink=0.05, width=0.5), fontsize=12, color='indianred')

# Customizing the layout and displaying the plot with the adjusted annotation for mean accuracy
# plt.title('FL Training Time and Test Accuracy for Various Cloud Configurations')
plt.tight_layout()
plt.show()
```
